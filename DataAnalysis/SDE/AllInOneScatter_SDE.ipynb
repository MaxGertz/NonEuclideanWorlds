```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

from mpl_toolkits.mplot3d import Axes3D
```

```python
all_files = glob.glob("data" + "/*.csv")
```

```python
datasets = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=False, header=0, sep=';')
    datasets.append(df)
```

```python
datasets[0]
```

```python
df = pd.concat(datasets).reset_index(drop=True)
```

```python
df
```

```python
datasets_player = []
datasets_constants = []
for i, row in df.iterrows():
    if row['gameobject_name'] == 'Player':
        datasets_player.append(row)
    else:
        datasets_constants.append(row)
```

```python
df_player = pd.DataFrame(datasets_player)
```

```python
df_player
```

```python
ax1 = df_player.plot.scatter(x='pos_x',y='pos_z')
ax1.set_title('Player movement')
ax1.set_xlabel('x in m')
ax1.set_ylabel('z in m')

#set axes limits
ax1.set_xlim([-3,5])
ax1.set_ylim([-3,3])
```
