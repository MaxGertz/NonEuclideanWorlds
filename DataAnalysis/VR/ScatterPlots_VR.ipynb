```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

from mpl_toolkits.mplot3d import Axes3D
```

```python
all_files = glob.glob("data" + "/*.csv")
```

```python
datasets = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=False, header=0, sep=';')
    datasets.append(df)
```

```python
ids = []
for filename in all_files:
    id = filename.split('_')
    ids.append(id[2].replace('study', 'p'))
```

```python
j = 0
for i in range(len(datasets)):
    df = datasets[i]
    datasets_player = []
    for i, row in df.iterrows():
        if row['gameobject_name'] == 'LeftEyeAnchor':
            datasets_player.append(row)
   
    df_player = pd.DataFrame(datasets_player)
    plot = df_player.plot.scatter(x='pos_x',y='pos_z')
    plot.set_title(ids[j])
    plot.set_xlabel('Width')
    plot.set_ylabel('Length')

    #set axes limits
    plot.set_xlim([-3,5])
    plot.set_ylim([-3,3])
    
    fig = plot.get_figure()
    fig.savefig("Maps\scatterplot_" + ids[j] + ".png", dpi=900)
    j = j+1
```
